```python
import pandas as pd
```

```python
df = pd.read_csv('3.Car info One hot encoding.csv')
df
```

# Milege, sell price and Age ,these are numerical form but car model has string,so we have to convert string into numerical form

```python
dummies = pd.get_dummies(df['Car Model'])
dummies
```

```python
merged =pd.concat([df,dummies],axis='columns')
merged
```

```python
#all data store in merged now delet or drop car model column
final =merged.drop(['Car Model'],axis='columns')
final
```

```python
#only dependent value is sell price so
y =df['Sell Price']
y
```

```python
#now all value in final so
x= final.drop(['Sell Price'],axis=1)
x
```

# complete seperate x and y so now use linear regression

```python
from sklearn.linear_model import LinearRegression
```

```python
model =LinearRegression()
model
```

```python
#fit then predict
model.fit(x,y)
```

```python
model.score(x,y)
```

```python
 # 1 for audi so others are zero
model.predict([[45000,8,1,0,0,0]])
```
